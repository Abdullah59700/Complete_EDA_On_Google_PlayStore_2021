```python
# Inspect the unique values
print(df_1['Size'].unique())

# Define the conversion function
def convert_size(size_str):
    if isinstance(size_str, float):  # Check if size_str is already a float
        return size_str  # Return the float value directly
    if size_str == 'Varies with device':
        return np.nan  # Handle 'Varies with device' as NaN
    
    size_str = size_str.replace(',', '').strip()  # Remove commas and strip whitespace
    if 'M' in size_str:
        return float(size_str.replace('M', '').strip()) * 1024 * 1024  # Convert MB to bytes
    elif 'K' in size_str:
        return float(size_str.replace('K', '').strip()) * 1024  # Convert KB to bytes
    elif 'G' in size_str:
        return float(size_str.replace('G', '').strip()) * 1024 * 1024 * 1024  # Convert GB to bytes
    else:
        return np.nan  # Handle any unexpected format as NaN

# Apply the function to the Size column
df_1['Size'] = df_1['Size'].apply(convert_size)

# Check the results
print(df_1['Size'].head())

# Handle NaN values (optional)
df_1.dropna(subset=['Size'], inplace=True)  # Option: drop rows with NaN in Size
```
